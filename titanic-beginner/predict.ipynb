```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("data/train.csv")
train.head()
```

```python
test = pd.read_csv("data/test.csv")
test.head()
```

```python
print(train.shape)
train.describe()
```

```python
print(test.shape)
test.describe()
```

```python
#欠損値の確認
def deficitTable(df):
    null_sum = df.isnull().sum()
    percent = 100 * null_sum / len(df)
    deficitTable = pd.concat([null_sum, percent], axis=1)
    deficitTable_renColumns = deficitTable.rename(
    columns = {0: "欠損値", 1: "%"}
    )
    return deficitTable_renColumns
```

```python
deficitTable(train)
```

```python
deficitTable(test)
```

```python
#訓練データの事前処理
#欠損データ補完
train["Age"] = train["Age"].fillna(train["Age"].median())
train["Embarked"] = train["Embarked"].fillna(train["Embarked"].mode()[0])

#文字列データ　-> 数値データ
train["Sex"][train["Sex"] == "male"] = 0
train["Sex"][train["Sex"] == "female"] = 1
train["Embarked"][train["Embarked"] == "S"] = 0
train["Embarked"][train["Embarked"] == "C"] = 1
train["Embarked"][train["Embarked"] == "Q"] = 2

train
```

```python
#検証データの事前処理
test["Age"] = test["Age"].fillna(test["Age"].median())
test["Embarked"] = test["Embarked"].fillna(test["Embarked"].mode()[0])
test["Fare"] = test["Fare"].fillna(test["Fare"].median())

test["Sex"][test["Sex"] == "male"] = 0
test["Sex"][test["Sex"] == "female"] = 1
test["Embarked"][test["Embarked"] == "S"] = 0
test["Embarked"][test["Embarked"] == "C"] = 1
test["Embarked"][test["Embarked"] == "Q"] = 2

test
```

# 学習モデルの作成
- 決定木

```python
from sklearn import tree 
```

```python
#目的変数と説明変数を定義
target = train["Survived"].values
features_one = train[["Pclass", "Sex", "Age", "Fare"]]

#決定木の作成
my_tree_one = tree.DecisionTreeClassifier()
my_tree_one = my_tree_one.fit(features_one, target)

#検証データの説明変数を取得
test_features = test[["Pclass", "Sex", "Age", "Fare"]]

#決定木を用いて、検証データを比較
my_prediction = my_tree_one.predict(test_features)
```

# csv出力

```python
#PassengerIdの取得
PassengerId = np.array(test["PassengerId"]).astype(int)

#予測データとPassengerIdをデータフレームに落とし込む
my_solution = pd.DataFrame(my_prediction, PassengerId, columns=["Survived"])

#my_tree_one.csvとして出力
my_solution.to_csv("my_tree_one.csv", index_label=["PassengerId"])
```

# 予測モデル2

```python
#予測モデル2で使うデータを取り出す
features_two = train[["Pclass", "Sex", "Age", "Fare", "SibSp", "Parch", "Embarked"]]

#決定木とアーギュメントの作成
#分類回数
max_depth = 10
#最小分割数(これを下回ったら以降対象ノードについて処理を行わない)
min_samples_split = 10
my_tree_two = tree.DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split)
my_tree_two = my_tree_two.fit(features_two, target)
```

```python
#testから予測モデル2で使うデータを取り出す
test_features2 = test[["Pclass", "Sex", "Age", "Fare", "SibSp", "Parch", "Embarked"]].values

#予測モデル2を用いてCSV書き込み
my_prediction2 = my_tree_two.predict(test_features2)
my_solution2 = pd.DataFrame(my_prediction2, PassengerId, columns=["Survived"])
my_solution2.to_csv("my_tree_two.csv", index_label="PassengerId")
```
